```python
pre_df = train_df.copy()


import numpy as np
pre_df['label'] = np.where(pre_df['label']== 'correct', 1, 0)
```

```python
distance = []
for index, row in train_df.iterrows():
    coords_1 = (row['pick_lat'], row['pick_lon'])
    coords_2 = (row['drop_lat'], row['drop_lon'])
    dis = geopy.distance.distance(coords_1, coords_2).km
    distance.append(dis)

```

```python
df1 = pre_df.pop('label') # remove column b and store it in df1
pre_df = pre_df.assign(dist=pd.Series(distance).values)
pre_df['label'] = df1
```

```python
pre_df = pre_df.dropna()
pre_df.head(20)
```

```python
selected_features = ['additional_fare', 'duration', 'meter_waiting', 'meter_waiting_fare', 'meter_waiting_till_pickup', 'fare', 'dist', 'label']

filtered_df = pre_df[selected_features]
filtered_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(filtered_df.drop('label',axis=1), 
           filtered_df['label'], test_size=0.25, random_state=10)


# X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(pre_df.drop(['label', 'pickup_time', 'drop_time', 'tripid'],axis=1), 
#            pre_df['label'], test_size=0.30, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification


clf = RandomForestClassifier(n_estimators = 1000, oob_score = True ,n_jobs = 1,random_state =1, 
                             class_weight="balanced", bootstrap=True, max_depth=15, 
                             max_features=4, min_samples_leaf=4, min_samples_split=5)

clf.fit(X_train, y_train)
y_pred_random_forest = clf.predict(X_test)
print(classification_report(y_test,y_pred_random_forest))

print(metrics.f1_score(y_test, y_pred_random_forest, average='macro'))
```
